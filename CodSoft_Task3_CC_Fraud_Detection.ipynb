Importing Dependencies That Are Required

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/drive/MyDrive/Dataset_csv/creditcard.csv')
credit_card_data.head(15)
```

```python
credit_card_data.tail()
```

```python
#Dataset Informations
credit_card_data.info()
```

```python
#To check the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
#Distribution of authorized transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

The Dataset is not balanced properly.
0 = Authorized Transaction
1 = Fraudulent Transaction

```python
#Separation of data for analysis
authorized = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(authorized.shape)
print(fraud.shape)
```

Statistical Measures of Authorized & Fraudulent Data.

```python
authorized.Amount.describe()
```

```python
fraud.Amount.describe()
```

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions.

Number of Fraudulent Transactions = 492

```python
authorized_sample = authorized.sample(n=492)
```

```python
#Concatenating Two DataFrames
new_dataset = pd.concat([authorized_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting of data into Features & Targets.

```python
#Data Splitting
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Training Data & Testing Data.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Training Model with Logistic Regression
model = LogisticRegression()
model.fit(X_train, Y_train)
```

Accuracy Score & Model Evaluation.

```python
#Accuracy of training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
#Accuracy of test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

*This Concludes that the model trained using creditcard dataset can detect the fradulent transactions.*
